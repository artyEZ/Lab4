```python
import pandas as pd
import datetime

df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.rename(columns={"Day": "day", "Exchange rate": "exchange_rate"}, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['deviation_from_median'] = abs(df['exchange_rate'] - df['exchange_rate'].median())
df['deviation_from_avarege_value'] = abs(df['exchange_rate'] - df['exchange_rate'].mean())
df
```

```python
def filter_by_avg(pd: pd.DataFrame, value: int) -> pd.DataFrame:
    """Write a function that takes as input a DataFrame 
    and a value of deviation from the average rate value, 
    and returns a DataFrame filtered by the value of 
    deviation from the average rate value.

    Args:
        pd (pd.DataFrame): start DataFrame
        average_value (int): deviation from the average rate value

    Returns:
        pd.DataFrame: filtered DataFrame
    """
    pf = df[df['deviation_from_avarege_value'] >= value]
    return pf

filter_by_avg(pd, df['exchange_rate'].mean())
```

```python
def sorted_by_dates(pd: pd.DataFrame, data_from: str, data_to: str) -> pd.DataFrame:
    """Write a function that takes a DataFrame, a start date and an end date as input, and returns a DataFrame filtered by dates.

    Args:
        pd (pd.DataFrame): Start DataFrame
        data_from (str): First date
        data_to (str): Second date

    Returns:
        pd.DataFrame: filtered DataFrame
    """
    df['day'] = pd.to_datetime(df['day'], format='%Y-%m')
    pf = df.loc[(df['day'] <= data_from) & (df['day'] >= data_to)]
    return pf
    
sorted_by_dates(pd, '2022-9-15', '2020-02-02')
```

```python
from typing import Tuple

def sorted_by_month(pd: pd.DataFrame, month: int) -> Tuple[pd.DataFrame, int]:
    """Group the DataFrame by month and calculate the average value of the rate.

    Args:
        pd (pd.DataFrame): Start dataframe
        month (int): Number of month

    Returns:
        Union[pd.DataFrame, int]: filtered DataFrame and deviation from the average rate value
    """
    df['month'] = df['day'].dt.month
    pf = df[df['month'] == month]
    return pf, pf['exchange_rate'].mean()

sorted_by_month(pd, 2)
```

```python
import datetime
def draw_weeks_stat(pd: pd.DataFrame, month: int, year: int) -> None:
    df['year'] = df['day'].dt.year
    pf = df[df['year'] == year]
    pf['month'] = df['day'].dt.month
    sf = pf[pf['year'] == year]

    list1=[]
    list2=[]

    x =  pf['exchange_rate']
    x.reset_index(drop=True, inplace=True)
    l = len(x)

    y1 = sf['exchange_rate'].mean()
    y2 = sf ['exchange_rate'].median()

    for i in range(l):
        list1.append(y1)
        list2.append(y2)     
   
    plt.ylabel('Exchange rate')
    plt.xlabel('Day')
    plt.plot(x, color='blue', linestyle='-.', linewidth=1, label='Exchange rate') 
    plt.plot(list1, color='green', linestyle='-', linewidth=1, label='Deviation from avg') 
    plt.plot(list2, color='red', linestyle='--', linewidth=1, label='Deviation from median')
    plt.legend()
    plt.title('Сurrency exchange rate changes over all time')
    plt.show()

draw_weeks_stat(df, 12, 1993)
```
